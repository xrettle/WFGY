````python
import os, json, zipfile, hashlib, csv, textwrap
from pathlib import Path
import pandas as pd

ZIP_NAME = 'PP02D_A_API_FULL_216_SHARDED_GREEN_PP02D_A_API_MVP_20260504T150642Z.zip'
zip_path = Path(ZIP_NAME)
if not zip_path.exists():
    try:
        from google.colab import files
        uploaded = files.upload()
        if ZIP_NAME not in uploaded:
            raise FileNotFoundError(f'Expected {ZIP_NAME}, but uploaded: {list(uploaded)}')
        zip_path = Path(ZIP_NAME)
    except Exception as exc:
        raise FileNotFoundError(f'Place {ZIP_NAME} next to this notebook or upload it in Colab.') from exc

extract_dir = Path('dd02a_extracted_results')
extract_dir.mkdir(exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as zf:
    zf.extractall(extract_dir)
print('Extracted files:', len(list(extract_dir.iterdir())))
````

````python
def load_json(name):
    with open(extract_dir / name, 'r', encoding='utf-8') as f:
        return json.load(f)

score = load_json('api_score_summary.json')
integrity = load_json('api_package_integrity_audit.json')
preflight = load_json('preflight_gate_record.json')
run_config = load_json('run_config.json')

summary_rows = [
    ('Run mode', score.get('api_run_mode')),
    ('Fixture count', score.get('fixture_count')),
    ('Shard count', score.get('shard_count')),
    ('API exceptions', score.get('api_exception_count')),
    ('Shard parse failures', score.get('shard_parse_fail_count')),
    ('Parse failures', score.get('parse_fail_count')),
    ('Fixture pass', score.get('fixture_pass_count')),
    ('Semantic citation checks', score.get('semantic_citation_checks')),
    ('Aggregate support union checks', score.get('aggregate_support_union_checks')),
    ('Total support closure', score.get('total_support_closure')),
    ('Counterfactual pairs', score.get('counterfactual_pairs_pass')),
    ('Poison rejection', preflight.get('poison_reject_count')),
    ('Sandbox A', preflight.get('sandbox_pass_a', {}).get('status')),
    ('Sandbox B', preflight.get('sandbox_pass_b', {}).get('status')),
    ('Package self-check', integrity.get('zip_self_check_pass')),
    ('Final verdict', score.get('api_verdict')),
]
summary_df = pd.DataFrame(summary_rows, columns=['Metric', 'Result'])
summary_df
````

````python
sandbox_df = pd.read_csv(extract_dir / 'full216_sharded_sandbox_ab_results.csv')
poison_df = pd.read_csv(extract_dir / 'full216_sharded_parser_poison_results.csv')
parse_df = pd.read_csv(extract_dir / 'shard_parse_trace.csv')
budget_df = pd.read_csv(extract_dir / 'shard_schema_budget_audit.csv')
display(sandbox_df)
display(poison_df)
display(parse_df)
display(budget_df)
````

# Final Polaris Protocol report

This final section generates a compact report for **PP02D_A — API Stability 216 Fixtures**.

It produces:
- `polaris_report_PP02D_A/executive_report.md`
- `polaris_report_PP02D_A/metrics_summary.csv`
- `polaris_report_PP02D_A/chart_*.png`

The charts are intentionally simple. They are meant to make the experiment result easy to inspect, not to expand the claim boundary.


````python
# Final Polaris Protocol report generator
# This cell uses the published result summary as a stable reporting fallback.
# If you adapt the notebook and compute fresh metrics, update POLARIS_PUBLISHED_METRICS before running this cell.

from pathlib import Path
import json, csv, math, re
import pandas as pd
import matplotlib.pyplot as plt

POLARIS_REPORT_EXPERIMENT_ID = "PP02D_A"
POLARIS_REPORT_TITLE = "PP02D_A — API Stability 216 Fixtures"
POLARIS_REPORT_REPO = "https://github.com/onestardao/WFGY"
POLARIS_REPORT_PURPOSE = "Reviews structured-output preservation and support closure across a 216-fixture, 6-shard API-style validation path."
POLARIS_REPORT_SPIRIT = "API stability supplement. This checks structured-output preservation, not real-world API guarantees or global QA improvement."
POLARIS_REPORT_CLAIM_BOUNDARY = "Structured API-style stability supplement only. Not proof that all API usage is stable."
POLARIS_PUBLISHED_SUMMARY = [
  "216 fixtures; 6 shards.",
  "0 API exceptions; 0 parse failures.",
  "216/216 fixture pass.",
  "1944/1944 semantic support checks; 2160/2160 support closure checks; 12/12 poison reject."
]
POLARIS_PUBLISHED_METRICS = {
  "Fixtures": 216,
  "Shards": 6,
  "API exceptions": 0,
  "Parse failures": 0,
  "Fixture pass": 216,
  "Fixture total": 216,
  "Semantic support checks": 1944,
  "Semantic support total": 1944,
  "Support closure checks": 2160,
  "Support closure total": 2160,
  "Poison reject": 12,
  "Poison total": 12
}
POLARIS_CHART_SPECS = [
  {
    "title": "Fixture pass",
    "ylabel": "Fixtures",
    "labels": [
      "Pass",
      "Total"
    ],
    "values": [
      216,
      216
    ]
  },
  {
    "title": "Support checks",
    "ylabel": "Checks",
    "labels": [
      "Semantic support",
      "Support closure"
    ],
    "values": [
      1944,
      2160
    ]
  },
  {
    "title": "Poison rejection",
    "ylabel": "Count",
    "labels": [
      "Rejected",
      "Total"
    ],
    "values": [
      12,
      12
    ]
  }
]

report_dir = Path(f"polaris_report_{POLARIS_REPORT_EXPERIMENT_ID}")
report_dir.mkdir(exist_ok=True)

# Save metrics table.
metrics_df = pd.DataFrame([
    {"metric": key, "value": value}
    for key, value in POLARIS_PUBLISHED_METRICS.items()
])
metrics_path = report_dir / "metrics_summary.csv"
metrics_df.to_csv(metrics_path, index=False)

# Generate compact charts.
chart_paths = []
def _safe_chart_name(title):
    return re.sub(r"[^a-z0-9_\\-]+", "_", title.lower().replace(" ", "_")).strip("_")

for idx, spec in enumerate(POLARIS_CHART_SPECS, start=1):
    labels = spec["labels"]
    values = spec["values"]
    plt.figure(figsize=(max(6, len(labels) * 1.2), 4))
    bars = plt.bar(labels, values)
    plt.title(spec["title"])
    plt.ylabel(spec.get("ylabel", "Value"))
    plt.xticks(rotation=25, ha="right")

    numeric_values = [float(v) for v in values if isinstance(v, (int, float))]
    max_value = max(numeric_values) if numeric_values else 0.0
    min_value = min(numeric_values) if numeric_values else 0.0

    if max_value == 0 and min_value == 0:
        # Make zero-count success charts visually readable instead of looking blank.
        plt.ylim(0, 1)
        for bar, value in zip(bars, values):
            x = bar.get_x() + bar.get_width() / 2
            plt.text(x, 0.05, str(value), ha="center", va="bottom", fontsize=10)
        plt.text(
            0.5, 0.88,
            "All tracked counts are 0",
            transform=plt.gca().transAxes,
            ha="center",
            va="center",
            fontsize=11,
        )
    else:
        upper = max_value * 1.18 if max_value > 0 else 1
        if min_value >= 0:
            plt.ylim(0, upper)
        for bar, value in zip(bars, values):
            x = bar.get_x() + bar.get_width() / 2
            y = bar.get_height()
            label = f"{value:.4g}" if isinstance(value, float) else str(value)
            plt.text(x, y + (upper * 0.02 if max_value > 0 else 0.03), label, ha="center", va="bottom", fontsize=9)

    plt.tight_layout()
    chart_path = report_dir / f"chart_{idx:02d}_{_safe_chart_name(spec['title'])}.png"
    plt.savefig(chart_path, dpi=180, bbox_inches="tight")
    plt.show()
    chart_paths.append(chart_path)

# Save executive markdown report.
summary_lines = "\n".join([f"- {item}" for item in POLARIS_PUBLISHED_SUMMARY])
chart_lines = "\n".join([f"- {p.name}" for p in chart_paths])

report_md = f"""# {POLARIS_REPORT_TITLE}

**Repository:** {POLARIS_REPORT_REPO}  
**Experiment ID:** {POLARIS_REPORT_EXPERIMENT_ID}

## Experiment spirit

{POLARIS_REPORT_SPIRIT}

## What this tests

{POLARIS_REPORT_PURPOSE}

## Published result summary

{summary_lines}

## Generated files

- metrics_summary.csv
{chart_lines}

## Claim boundary

{POLARIS_REPORT_CLAIM_BOUNDARY}

This report is designed for reproduction / inspection. It does not convert scoped evidence into universal proof.
"""

report_path = report_dir / "executive_report.md"
report_path.write_text(report_md, encoding="utf-8")

print("Polaris Protocol report generated:")
print(f"- {report_path}")
print(f"- {metrics_path}")
for path in chart_paths:
    print(f"- {path}")
````
