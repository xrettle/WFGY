````python
# =========================
# CASE GENERATION
# =========================
# Public-prompt safety:
# - Prompts include public target handles and candidate certificate.
# - Prompts do NOT include expected_validity or valid-set labels.
# - Expected validity is stored only for scorer after API output.

VALIDITY_LAW = """
You are auditing exactly one certificate.

Return only this exact JSON object shape:
{"certificate_handle":"<same certificate_handle>","certificate_validity":"VALID|INVALID","confidence_policy":"NO_FREE_CLAIM"}

No markdown. No explanation. No extra fields. No arrays.

Decision law:
A certificate is VALID iff all seven locks pass. If any lock fails, output INVALID.

L0 Terminal answer binding:
  certificate.answer_candidate_handle must equal TARGET.required_answer_candidate_handle.

L1 Edge exact:
  certificate.proof_edge_handles must be exactly the duplicate-free set TARGET.required_edge_handles.
  Order does not matter. No missing handles. No extra handles. No duplicates.

L2 Edge bound:
  every proof edge in certificate.proof_edge_handles must be listed as BOUND in PUBLIC_EDGE_BINDINGS.

L3 Source support exact:
  certificate.source_span_handles must be exactly the duplicate-free set TARGET.required_source_span_handles.
  Order does not matter. No missing handles. No extra handles. No duplicates.

L4 Source authority:
  every source span in certificate.source_span_handles must be listed as AUTHORITATIVE in PUBLIC_SOURCE_REGISTRY.

L5 Distractor rejection exact:
  certificate.rejected_distractor_handles must be exactly the duplicate-free set TARGET.required_rejected_distractor_handles.
  Order does not matter. No missing handles. No extra handles. No duplicates.

L6 Claim allowed:
  certificate.claim_ceiling must equal TARGET.required_claim_ceiling.

Important positive rule:
  If L0, L1, L2, L3, L4, L5, and L6 all pass, you MUST output VALID.
  Do not over-reject a valid certificate.

Important negative rule:
  If the answer is wrong, output INVALID even if all route/topology/source/distractor/claim fields look complete.
""".strip()

INVALID_FAMILIES = [
    "wrong_answer_complete_topology",
    "missing_edge",
    "extra_edge",
    "duplicate_edge",
    "missing_source_span",
    "extra_source_span",
    "duplicate_source_span",
    "non_authoritative_source",
    "missing_distractor_rejection",
    "extra_distractor_rejection",
    "duplicate_distractor_rejection",
    "bad_claim_ceiling",
    "terminal_answer_only_bad_topology",
    "wrong_answer_plus_extra_support",
    "wrong_answer_plus_duplicate_distractor",
]

BUCKET_DISTRIBUTION = {0: 11, 1: 11, 2: 10, 3: 11, 4: 11}  # 54 arenas, 108 valid / 108 invalid

def make_public_world(i: int):
    p = f"{i:03d}"
    return {
        "arena_handle": f"AR_FINAL_{p}",
        "answer_candidate_table": {
            f"A_{p}_T": f"Vault_{p}_TERMINAL",
            f"A_{p}_W": f"Gate_{p}_RUMOR",
            f"A_{p}_Z": f"Harbor_{p}_DISTRACTOR",
        },
        "target": {
            "required_answer_candidate_handle": f"A_{p}_T",
            "required_edge_handles": [f"E_{p}_A", f"E_{p}_B"],
            "required_source_span_handles": [f"S_{p}_A", f"S_{p}_B"],
            "required_rejected_distractor_handles": [f"D_{p}_A"],
            "required_claim_ceiling": "SOURCE_GROUNDED_SCOPED",
        },
        "public_edge_bindings": {
            f"E_{p}_A": "BOUND",
            f"E_{p}_B": "BOUND",
            f"E_{p}_X": "UNBOUND",
        },
        "public_source_registry": {
            f"S_{p}_A": "AUTHORITATIVE",
            f"S_{p}_B": "AUTHORITATIVE",
            f"S_{p}_C": "NON_AUTHORITATIVE",
            f"S_{p}_EXTRA": "AUTHORITATIVE",
        },
        "public_world_text": [
            f"S_{p}_A says the object travels from Node_{p}_START to Relay_{p}_MID via E_{p}_A.",
            f"S_{p}_B says Relay_{p}_MID terminates at Vault_{p}_TERMINAL via E_{p}_B.",
            f"S_{p}_C is a non-authoritative rumor saying it terminates at Gate_{p}_RUMOR."
        ],
    }

def valid_certificate(world, cert_handle):
    t = world["target"]
    return {
        "certificate_handle": cert_handle,
        "answer_candidate_handle": t["required_answer_candidate_handle"],
        "proof_edge_handles": list(t["required_edge_handles"]),
        "source_span_handles": list(t["required_source_span_handles"]),
        "rejected_distractor_handles": list(t["required_rejected_distractor_handles"]),
        "claim_ceiling": t["required_claim_ceiling"],
    }

def invalid_certificate(world, cert_handle, family):
    c = valid_certificate(world, cert_handle)
    t = world["target"]
    p = world["arena_handle"].split("_")[-1]
    if family == "wrong_answer_complete_topology":
        c["answer_candidate_handle"] = f"A_{p}_W"
    elif family == "missing_edge":
        c["proof_edge_handles"] = [t["required_edge_handles"][0]]
    elif family == "extra_edge":
        c["proof_edge_handles"] = t["required_edge_handles"] + [f"E_{p}_X"]
    elif family == "duplicate_edge":
        c["proof_edge_handles"] = [t["required_edge_handles"][0], t["required_edge_handles"][0], t["required_edge_handles"][1]]
    elif family == "missing_source_span":
        c["source_span_handles"] = [t["required_source_span_handles"][1]]
    elif family == "extra_source_span":
        c["source_span_handles"] = t["required_source_span_handles"] + [f"S_{p}_EXTRA"]
    elif family == "duplicate_source_span":
        c["source_span_handles"] = [t["required_source_span_handles"][0], t["required_source_span_handles"][0], t["required_source_span_handles"][1]]
    elif family == "non_authoritative_source":
        c["source_span_handles"] = [t["required_source_span_handles"][0], f"S_{p}_C"]
    elif family == "missing_distractor_rejection":
        c["rejected_distractor_handles"] = []
    elif family == "extra_distractor_rejection":
        c["rejected_distractor_handles"] = t["required_rejected_distractor_handles"] + [f"D_{p}_EXTRA"]
    elif family == "duplicate_distractor_rejection":
        c["rejected_distractor_handles"] = [t["required_rejected_distractor_handles"][0], t["required_rejected_distractor_handles"][0]]
    elif family == "bad_claim_ceiling":
        c["claim_ceiling"] = "FREE_CLAIM"
    elif family == "terminal_answer_only_bad_topology":
        c["proof_edge_handles"] = []
        c["source_span_handles"] = []
        c["rejected_distractor_handles"] = []
    elif family == "wrong_answer_plus_extra_support":
        c["answer_candidate_handle"] = f"A_{p}_W"
        c["source_span_handles"] = t["required_source_span_handles"] + [f"S_{p}_EXTRA"]
    elif family == "wrong_answer_plus_duplicate_distractor":
        c["answer_candidate_handle"] = f"A_{p}_W"
        c["rejected_distractor_handles"] = [t["required_rejected_distractor_handles"][0], t["required_rejected_distractor_handles"][0]]
    else:
        raise ValueError(f"Unknown family: {family}")
    return c

def compile_validity(world, cert):
    t = world["target"]
    edge_bindings = world["public_edge_bindings"]
    source_registry = world["public_source_registry"]

    def exact_dup_free(observed, required):
        return (
            isinstance(observed, list)
            and len(observed) == len(set(observed))
            and set(observed) == set(required)
        )

    L0 = cert.get("answer_candidate_handle") == t["required_answer_candidate_handle"]
    L1 = exact_dup_free(cert.get("proof_edge_handles"), t["required_edge_handles"])
    L2 = L1 and all(edge_bindings.get(e) == "BOUND" for e in cert.get("proof_edge_handles", []))
    L3 = exact_dup_free(cert.get("source_span_handles"), t["required_source_span_handles"])
    L4 = L3 and all(source_registry.get(s) == "AUTHORITATIVE" for s in cert.get("source_span_handles", []))
    L5 = exact_dup_free(cert.get("rejected_distractor_handles"), t["required_rejected_distractor_handles"])
    L6 = cert.get("claim_ceiling") == t["required_claim_ceiling"]
    locks = {"L0": L0, "L1": L1, "L2": L2, "L3": L3, "L4": L4, "L5": L5, "L6": L6}
    return ("VALID" if all(locks.values()) else "INVALID"), locks

def build_prompt(world, cert):
    public_packet = {
        "AUDIT_TARGETS": world["target"],
        "ANSWER_CANDIDATE_TABLE": world["answer_candidate_table"],
        "PUBLIC_EDGE_BINDINGS": world["public_edge_bindings"],
        "PUBLIC_SOURCE_REGISTRY": world["public_source_registry"],
        "PUBLIC_WORLD_TEXT": world["public_world_text"],
        "CERTIFICATE_TO_AUDIT": cert,
    }
    return (
        VALIDITY_LAW
        + "\\n\\nPUBLIC_PACKET:\\n"
        + json.dumps(public_packet, ensure_ascii=False, indent=2)
        + "\\n\\nReturn only the JSON verdict object now."
    )

rows = []
arenas = []
family_counter = 0
arena_index = 0
for k, arena_count in BUCKET_DISTRIBUTION.items():
    for _ in range(arena_count):
        arena_index += 1
        world = make_public_world(arena_index)
        cert_rows = []
        valid_positions = set(range(k))  # first k valid; hidden from prompt, used only for construction
        for slot in range(4):
            cert_handle = f"C_FINAL_{arena_index:03d}_{chr(ord('A')+slot)}"
            if slot in valid_positions:
                family = "valid_control"
                cert = valid_certificate(world, cert_handle)
            else:
                family = INVALID_FAMILIES[family_counter % len(INVALID_FAMILIES)]
                family_counter += 1
                cert = invalid_certificate(world, cert_handle, family)
            expected, locks = compile_validity(world, cert)
            prompt_id = f"P_FINAL_{arena_index:03d}_{chr(ord('A')+slot)}"
            row = {
                "row_key": f"ROW_FINAL_{arena_index:03d}_{chr(ord('A')+slot)}",
                "prompt_id": prompt_id,
                "arena_handle": world["arena_handle"],
                "certificate_handle": cert_handle,
                "certificate": cert,
                "public_world": world,
                "prompt": build_prompt(world, cert),
                "expected_validity": expected,
                "lock_vector": locks,
                "family": family,
                "valid_set_size_bucket": k,
            }
            cert_rows.append(row)
            rows.append(row)
        arenas.append({"arena_handle": world["arena_handle"], "valid_set_size_bucket": k})

assert len(rows) == 216, len(rows)
assert sum(1 for r in rows if r["expected_validity"] == "VALID") == 108
assert sum(1 for r in rows if r["expected_validity"] == "INVALID") == 108
print("Generated rows:", len(rows))
print("Expected VALID:", sum(1 for r in rows if r["expected_validity"] == "VALID"))
print("Expected INVALID:", sum(1 for r in rows if r["expected_validity"] == "INVALID"))
print("Arenas:", len(arenas))
````

````python
# =========================
# STRICT PARSER V6
# =========================

class DuplicateKeyError(ValueError):
    pass

def no_duplicate_object_pairs_hook(pairs):
    obj = {}
    for k, v in pairs:
        if k in obj:
            raise DuplicateKeyError(f"Duplicate key: {k}")
        obj[k] = v
    return obj

ALLOWED_FIELDS = {"certificate_handle", "certificate_validity", "confidence_policy"}
VALIDITY_ENUM = {"VALID", "INVALID"}

def parse_model_output(raw_text, prompted_certificate_handle):
    rec = {
        "parse_status": None,
        "parser_rescue_used": False,
        "parse_error_code": None,
        "parsed_certificate_handle": None,
        "parsed_certificate_validity": None,
        "parsed_confidence_policy": None,
    }
    try:
        if raw_text is None:
            rec["parse_status"] = "PARSE_FAIL_EMPTY"
            rec["parse_error_code"] = "EMPTY_OUTPUT"
            return rec
        if not isinstance(raw_text, str):
            rec["parse_status"] = "PARSE_FAIL_BAD_RAW_TYPE"
            rec["parse_error_code"] = "RAW_OUTPUT_NOT_STRING"
            return rec
        s = raw_text.strip()
        if not s:
            rec["parse_status"] = "PARSE_FAIL_EMPTY"
            rec["parse_error_code"] = "EMPTY_OUTPUT"
            return rec
        if "```" in s:
            rec["parse_status"] = "PARSE_FAIL_MARKDOWN_FENCE"
            rec["parse_error_code"] = "MARKDOWN_FENCE"
            return rec
        if not (s.startswith("{") and s.endswith("}")):
            rec["parse_status"] = "PARSE_FAIL_EXTRA_TEXT_OR_NON_OBJECT"
            rec["parse_error_code"] = "TEXT_OUTSIDE_JSON_OR_NOT_OBJECT"
            return rec
        try:
            obj = json.loads(s, object_pairs_hook=no_duplicate_object_pairs_hook)
        except DuplicateKeyError:
            rec["parse_status"] = "PARSE_FAIL_DUPLICATE_KEY"
            rec["parse_error_code"] = "DUPLICATE_KEY"
            return rec
        except json.JSONDecodeError as e:
            rec["parse_status"] = "PARSE_FAIL_SYNTAX"
            rec["parse_error_code"] = f"JSON_SYNTAX:{str(e)[:120]}"
            return rec
        if not isinstance(obj, dict):
            rec["parse_status"] = "PARSE_FAIL_NOT_OBJECT"
            rec["parse_error_code"] = "JSON_NOT_OBJECT"
            return rec
        fields = set(obj.keys())
        if fields != ALLOWED_FIELDS:
            missing = sorted(ALLOWED_FIELDS - fields)
            extra = sorted(fields - ALLOWED_FIELDS)
            rec["parse_status"] = "PARSE_FAIL_FIELD_SET"
            rec["parse_error_code"] = f"MISSING={missing};EXTRA={extra}"
            return rec

        ch = obj.get("certificate_handle")
        cv = obj.get("certificate_validity")
        cp = obj.get("confidence_policy")
        if not isinstance(ch, str):
            rec["parse_status"] = "PARSE_FAIL_BAD_TYPE_CERTIFICATE_HANDLE"
            rec["parse_error_code"] = "CERTIFICATE_HANDLE_NOT_STRING"
            return rec
        if not isinstance(cv, str):
            rec["parse_status"] = "PARSE_FAIL_BAD_TYPE_CERTIFICATE_VALIDITY"
            rec["parse_error_code"] = "CERTIFICATE_VALIDITY_NOT_STRING"
            return rec
        if not isinstance(cp, str):
            rec["parse_status"] = "PARSE_FAIL_BAD_TYPE_CONFIDENCE_POLICY"
            rec["parse_error_code"] = "CONFIDENCE_POLICY_NOT_STRING"
            return rec
        if ch != prompted_certificate_handle:
            rec["parse_status"] = "PARSE_FAIL_HANDLE_MISMATCH"
            rec["parse_error_code"] = f"HANDLE_MISMATCH:{ch}!={prompted_certificate_handle}"
            return rec
        if cv not in VALIDITY_ENUM:
            rec["parse_status"] = "PARSE_FAIL_BAD_ENUM"
            rec["parse_error_code"] = f"BAD_VALIDITY_ENUM:{cv}"
            return rec
        if cp != "NO_FREE_CLAIM":
            rec["parse_status"] = "PARSE_FAIL_BAD_CONFIDENCE_POLICY"
            rec["parse_error_code"] = f"BAD_CONFIDENCE_POLICY:{cp}"
            return rec

        rec.update({
            "parse_status": "PARSE_PASS",
            "parsed_certificate_handle": ch,
            "parsed_certificate_validity": cv,
            "parsed_confidence_policy": cp,
        })
        return rec
    except Exception as e:
        # Controlled parser failure, never crash notebook without record.
        rec["parse_status"] = "PARSE_FAIL_INTERNAL_CONTROLLED"
        rec["parse_error_code"] = f"CONTROLLED_EXCEPTION:{type(e).__name__}:{str(e)[:120]}"
        return rec

print("Parser V6 loaded.")
````

````python
# =========================
# API KEY + CLIENT
# =========================

api_key = getpass("Paste OpenAI API key: ")
client = OpenAI(api_key=api_key)
print("Client initialized. Key is not printed or stored.")
````

````python
# =========================
# API RUN
# =========================

raw_records = []
transport_records = []
semantic_fallback_count = 0
parser_rescue_count = 0

def response_to_text(resp):
    # Prefer SDK convenience property.
    txt = getattr(resp, "output_text", None)
    if isinstance(txt, str) and txt.strip():
        return txt
    # Fallback extraction is transport-only, not semantic. It does not repair content.
    try:
        chunks = []
        for item in getattr(resp, "output", []) or []:
            for c in getattr(item, "content", []) or []:
                t = getattr(c, "text", None)
                if isinstance(t, str):
                    chunks.append(t)
        if len(chunks) == 1:
            return chunks[0]
        if len(chunks) > 1:
            raise RuntimeError("AMBIGUOUS_MULTIPLE_TEXT_CHUNKS")
    except Exception as e:
        raise
    return ""

for idx, row in enumerate(rows, start=1):
    prompt = row["prompt"]
    transport = {
        "row_key": row["row_key"],
        "prompt_id": row["prompt_id"],
        "arena_handle": row["arena_handle"],
        "certificate_handle": row["certificate_handle"],
        "transport_status": None,
        "attempt_count": 0,
        "error": None,
        "model": MODEL_NAME,
    }
    raw_text = None
    last_error = None
    for attempt in range(MAX_TRANSPORT_RETRIES + 1):
        transport["attempt_count"] += 1
        try:
            resp = client.responses.create(
                model=MODEL_NAME,
                input=prompt,
                max_output_tokens=MAX_OUTPUT_TOKENS,
            )
            raw_text = response_to_text(resp)
            transport["transport_status"] = "TRANSPORT_PASS"
            break
        except Exception as e:
            last_error = f"{type(e).__name__}: {str(e)[:300]}"
            if attempt < MAX_TRANSPORT_RETRIES:
                time.sleep(1.0 + attempt)
                continue
            transport["transport_status"] = "TRANSPORT_FAIL"
            transport["error"] = last_error
            raw_text = ""
    raw_records.append({
        "row_key": row["row_key"],
        "prompt_id": row["prompt_id"],
        "arena_handle": row["arena_handle"],
        "certificate_handle": row["certificate_handle"],
        "raw_output": raw_text,
        "raw_output_sha256": hashlib.sha256((raw_text or "").encode("utf-8")).hexdigest(),
    })
    transport_records.append(transport)
    print(f"[{idx:03d}/216] {row['certificate_handle']} {transport['transport_status']}")
    if REQUEST_SLEEP_SECONDS:
        time.sleep(REQUEST_SLEEP_SECONDS)

print("API run complete.")
````

````python
# =========================
# PARSE + SCORE
# =========================

raw_by_key = {r["row_key"]: r for r in raw_records}
parser_records = []
scorer_trace = []

for row in rows:
    raw_text = raw_by_key[row["row_key"]]["raw_output"]
    p = parse_model_output(raw_text, row["certificate_handle"])
    parser_records.append({
        "row_key": row["row_key"],
        "prompt_id": row["prompt_id"],
        "arena_handle": row["arena_handle"],
        "certificate_handle": row["certificate_handle"],
        "parse_status": p["parse_status"],
        "parser_rescue_used": p["parser_rescue_used"],
        "parse_error_code": p["parse_error_code"],
        "parsed_certificate_handle": p["parsed_certificate_handle"],
        "parsed_certificate_validity": p["parsed_certificate_validity"],
        "parsed_confidence_policy": p["parsed_confidence_policy"],
    })
    model_validity = p["parsed_certificate_validity"] if p["parse_status"] == "PARSE_PASS" else None
    exact = (model_validity == row["expected_validity"])
    scorer_trace.append({
        "row_key": row["row_key"],
        "prompt_id": row["prompt_id"],
        "arena_handle": row["arena_handle"],
        "certificate_handle": row["certificate_handle"],
        "family": row["family"],
        "valid_set_size_bucket": row["valid_set_size_bucket"],
        "expected_validity": row["expected_validity"],
        "model_validity": model_validity,
        "certificate_exact": exact,
        "parse_status": p["parse_status"],
    })

expected_valid_sets = defaultdict(set)
model_valid_sets = defaultdict(set)
for st in scorer_trace:
    if st["expected_validity"] == "VALID":
        expected_valid_sets[st["arena_handle"]].add(st["certificate_handle"])
    if st["model_validity"] == "VALID":
        model_valid_sets[st["arena_handle"]].add(st["certificate_handle"])

arena_summary = []
for a in sorted({r["arena_handle"] for r in rows}):
    exp = sorted(expected_valid_sets[a])
    mod = sorted(model_valid_sets[a])
    arena_summary.append({
        "arena_handle": a,
        "expected_valid_set": exp,
        "model_valid_set": mod,
        "arena_valid_set_match": exp == mod,
    })

# Metrics
api_calls = len(raw_records)
transport_pass = sum(1 for t in transport_records if t["transport_status"] == "TRANSPORT_PASS")
parse_pass = sum(1 for p in parser_records if p["parse_status"] == "PARSE_PASS")
parse_fail = len(parser_records) - parse_pass
cert_exact = sum(1 for s in scorer_trace if s["certificate_exact"])
arena_match = sum(1 for a in arena_summary if a["arena_valid_set_match"])
valid_total = sum(1 for r in rows if r["expected_validity"] == "VALID")
invalid_total = sum(1 for r in rows if r["expected_validity"] == "INVALID")
valid_recall = sum(1 for s in scorer_trace if s["expected_validity"] == "VALID" and s["model_validity"] == "VALID")
invalid_rejection = sum(1 for s in scorer_trace if s["expected_validity"] == "INVALID" and s["model_validity"] == "INVALID")
transport_fail = len(transport_records) - transport_pass
parser_rescue = sum(1 for p in parser_records if p["parser_rescue_used"])

bucket_summary = []
for bucket in sorted(set(r["valid_set_size_bucket"] for r in rows)):
    arena_handles = sorted({r["arena_handle"] for r in rows if r["valid_set_size_bucket"] == bucket})
    matches = sum(1 for a in arena_summary if a["arena_handle"] in arena_handles and a["arena_valid_set_match"])
    bucket_summary.append({
        "valid_set_size_bucket": bucket,
        "arena_count": len(arena_handles),
        "arena_match": matches,
        "required": len(arena_handles),
        "status": "PASS" if matches == len(arena_handles) else "FAIL",
    })

family_summary = []
for fam in sorted(set(r["family"] for r in rows)):
    fam_rows = [s for s in scorer_trace if s["family"] == fam]
    exact_count = sum(1 for s in fam_rows if s["certificate_exact"])
    family_summary.append({
        "family": fam,
        "rows": len(fam_rows),
        "certificate_exact": exact_count,
        "required": len(fam_rows),
        "status": "PASS" if exact_count == len(fam_rows) else "FAIL",
    })

green = (
    api_calls == 216 and
    transport_pass == 216 and
    parse_pass == 216 and
    parse_fail == 0 and
    parser_rescue == 0 and
    cert_exact == 216 and
    arena_match == 54 and
    valid_recall == valid_total == 108 and
    invalid_rejection == invalid_total == 108 and
    semantic_fallback_count == 0 and
    transport_fail == 0
)

verdict = {
    "run_id": RUN_ID,
    "run_type": "API_ONLY_FINAL_216",
    "artifact_role": "API_EVIDENCE_ZIP_NOT_PROOF",
    "model": MODEL_NAME,
    "api_calls": f"{api_calls}/216",
    "transport_pass": f"{transport_pass}/216",
    "transport_fail": transport_fail,
    "parse_pass": f"{parse_pass}/216",
    "parse_fail": parse_fail,
    "parser_rescue": parser_rescue,
    "semantic_fallback_count": semantic_fallback_count,
    "certificate_exact": f"{cert_exact}/216",
    "arena_valid_set_match": f"{arena_match}/54",
    "valid_recall": f"{valid_recall}/{valid_total}",
    "invalid_rejection": f"{invalid_rejection}/{invalid_total}",
    "run_disposition": "GREEN" if green else "RED",
    "claim_ceiling": "API_216_EVIDENCE_ONLY_NOT_GLOBAL_PROOF",
}

print(json.dumps(verdict, indent=2, ensure_ascii=False))

# Show failures
failed_rows = [s for s in scorer_trace if not s["certificate_exact"]]
failed_arenas = [a for a in arena_summary if not a["arena_valid_set_match"]]
print("Failed rows:", len(failed_rows))
for x in failed_rows[:20]:
    print(x)
print("Failed arenas:", len(failed_arenas))
for x in failed_arenas[:10]:
    print(x)
````

````python
# =========================
# WRITE EVIDENCE ZIP
# =========================

def write_jsonl(path, records):
    with open(path, "w", encoding="utf-8") as f:
        for r in records:
            f.write(json.dumps(r, ensure_ascii=False) + "\n")

def write_csv(path, records):
    records = list(records)
    if not records:
        Path(path).write_text("", encoding="utf-8")
        return
    # Serialize list/dict fields as JSON strings.
    keys = list(records[0].keys())
    with open(path, "w", encoding="utf-8", newline="") as f:
        w = csv.DictWriter(f, fieldnames=keys)
        w.writeheader()
        for r in records:
            rr = {}
            for k, v in r.items():
                rr[k] = json.dumps(v, ensure_ascii=False) if isinstance(v, (dict, list, set)) else v
            w.writerow(rr)

# Redact prompts from public rows? We include prompts as evidence.
case_public = []
for r in rows:
    case_public.append({
        "row_key": r["row_key"],
        "prompt_id": r["prompt_id"],
        "arena_handle": r["arena_handle"],
        "certificate_handle": r["certificate_handle"],
        "certificate": r["certificate"],
        "public_world": r["public_world"],
        "prompt": r["prompt"],
        "expected_validity": r["expected_validity"],
        "family": r["family"],
        "valid_set_size_bucket": r["valid_set_size_bucket"],
        "lock_vector": r["lock_vector"],
    })

write_jsonl(OUT / "public_prompts_and_cases.jsonl", case_public)
write_jsonl(OUT / "raw_outputs.jsonl", raw_records)
write_csv(OUT / "transport_records.csv", transport_records)
write_csv(OUT / "parser_records.csv", parser_records)
write_csv(OUT / "scorer_trace.csv", scorer_trace)
write_csv(OUT / "arena_valid_set_summary.csv", arena_summary)
write_csv(OUT / "bucket_summary.csv", bucket_summary)
write_csv(OUT / "family_summary.csv", family_summary)
(OUT / "verdict.json").write_text(json.dumps(verdict, indent=2, ensure_ascii=False), encoding="utf-8")
(OUT / "config.json").write_text(json.dumps({
    "run_id": RUN_ID,
    "model": MODEL_NAME,
    "max_output_tokens": MAX_OUTPUT_TOKENS,
    "request_sleep_seconds": REQUEST_SLEEP_SECONDS,
    "max_transport_retries": MAX_TRANSPORT_RETRIES,
    "planned_calls": 216,
    "arena_count": 54,
    "certificate_count": 216,
    "valid": 108,
    "invalid": 108,
    "parser": "V6 strict no-rescue",
    "fallback": "No semantic fallback; transport retry disabled by default",
}, indent=2, ensure_ascii=False), encoding="utf-8")

# README
(OUT / "README_EVIDENCE.md").write_text(f"""# PP02D-B R9 API-only FINAL 216 Evidence

Run ID: {RUN_ID}

This zip is evidence sediment, not proof or closure.

Required GREEN:
- transport pass 216/216
- parse pass 216/216
- certificate exact 216/216
- arena valid-set 54/54
- valid recall 108/108
- invalid rejection 108/108
- parser rescue 0
- semantic fallback 0

Verdict:
```json
{json.dumps(verdict, indent=2, ensure_ascii=False)}
```
""", encoding="utf-8")

# Hash manifest before zip
hash_rows = []
for p in sorted(OUT.glob("*")):
    if p.is_file():
        data = p.read_bytes()
        hash_rows.append({
            "relative_path": p.name,
            "file_size_bytes": len(data),
            "sha256": hashlib.sha256(data).hexdigest(),
        })
write_csv(OUT / "hash_manifest.csv", hash_rows)

# Zip
zip_path = Path("/content") / f"{RUN_ID}_EVIDENCE.zip"
with zipfile.ZipFile(zip_path, "w", compression=zipfile.ZIP_DEFLATED) as z:
    for p in sorted(OUT.glob("*")):
        z.write(p, arcname=p.name)

print("Evidence zip:", zip_path)
print("Verdict:", verdict["run_disposition"])

if IN_COLAB:
    files.download(str(zip_path))
else:
    print("Not in Colab; download manually:", zip_path)
````
